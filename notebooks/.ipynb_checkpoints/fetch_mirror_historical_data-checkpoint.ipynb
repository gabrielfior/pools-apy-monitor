```python
import requests
import json
import datetime
import sqlalchemy
import operator
import pandas as pd
import psycopg2
from collections import defaultdict
```

```python
url = 'http://www.mirrortracker.info/api/asset-stats/daily-average'
r = requests.get(url)
```

```python
data = r.json()
```

```python
len(data)
```

```python
data[0]
```

```python
d = defaultdict(list)
assets = {}
for i in data:
    d[i['asset_id']].append({'apy_long': i['avg_long'], 'apy_short':i['avg_short'], 
                             'date':datetime.datetime.strptime(i['stat_date'],'%m-%d-%Y')})
    if i['asset_id'] not in assets:
        assets[i['asset_id']] = i['asset']
```
